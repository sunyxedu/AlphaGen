```python
import tensorflow as tf
import numpy as np
import pickle
import os

# Define the loss function
def loss_func(y_true, y_pred):
    ignore_steps = 20
    y_true = tf.cast(y_true, tf.float32)
    
    time_steps = tf.shape(y_true)[1]
    
    mask = tf.reduce_all(tf.equal(y_true, -1), axis=-1)
    mask = tf.cast(mask, dtype=tf.float32)
    
    loss = tf.abs(y_pred[:, ignore_steps:, :] - y_true[:, ignore_steps:, :])
    loss = loss * (1 - mask[:, ignore_steps:])[:, :, None]
    return tf.reduce_mean(loss)

# Load the model
# with tf.keras.utils.custom_object_scope({'loss_func': loss_func}):
model = tf.keras.models.load_model('model_full_110epochs.h5')

model.summary()
```
