```python
import tensorflow as tf
import numpy as np
import pickle
import os

# Define the loss function
def cancel_loss(y_true, y_pred):
    y_true_truncated = y_true[:, 40:, :, :]
    y_pred_truncated = y_pred[:, 40:, :, :]
    loss = tf.keras.losses.categorical_crossentropy(y_true_truncated, y_pred_truncated)
    return tf.reduce_mean(loss)

# Load the model
with tf.keras.utils.custom_object_scope({'cancel_loss': cancel_loss}):
    model = tf.keras.models.load_model('model_full.h5')

model.summary()
```

```python
replay_path = '/Users/yuxuan/Documents/AlphaGen/Replays/Match_Data/MeltedToast/0_zBBlWiw.pkl'
feat = []
army = []
Y = []

with open(replay_path,"rb") as file:
    total = pickle.load(file)
    feat.append(np.array(total["features"]))
    army.append(np.array(total["armies"]))
    Y.append(np.array(total["kings"]))
print(len(feat[0]))
```
